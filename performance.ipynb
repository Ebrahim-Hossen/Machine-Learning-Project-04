```python
# Specify the file name
file_name3 = "/kaggle/input/test180611/resnet50_candidate.txt"  # Replace with the actual file name

# Initialize an empty list to store the lines
rgenerate = []

# Open the file for reading
with open(file_name3, "r") as file:
    # Read the file line by line
    for line in file:
        # Add single quotes around each line and append to the list
        rgenerate.append(f"'{line.strip()}'")
```

```python
# Specify the file name
file_name4 = "/kaggle/input/test180611/resnet50_reference.txt"  # Replace with the actual file name

# Initialize an empty list to store the lines
rreference = []

# Open the file for reading
with open(file_name4, "r") as file:
    # Read the file line by line
    for line in file:
        # Add single quotes around each line and append to the list
        rreference.append(f"{line.strip()}")
```

```python
print(rgenerate[0])
print(rreference[0])
```

```python
from nltk.util import ngrams
import numpy as np
# Reference and generated captions
# reference_caption = ['শূন্যে থাকা একজন বাস্কেটবল খেলোয়াড় বল শুট করতে নিচ্ছে']
# generated_caption = 'শূন্যে থাকা একজন বাস্কেটবল খেলোয়াড়'
cider_value=[]
cnt=0
for i in range(len(rreference)):
    reference_caption = [rreference[i]]
    generated_caption = rgenerate[i]
    # Tokenize the captions
    #print("ref ----- : ",reference_caption)
    #print("gen ----- : ",generated_caption)
    reference_tokens = nltk.word_tokenize(reference_caption[0])
    generated_tokens = nltk.word_tokenize(generated_caption)

# Calculate n-grams for reference and generated captions
    n = 1  # You can choose the n-gram order you want
    reference_ngrams = list(ngrams(reference_tokens, n))
    generated_ngrams = list(ngrams(generated_tokens, n))
# Calculate CIDEr score
    intersection_ngram_count = len(set(reference_ngrams) & set(generated_ngrams))
    reference_ngram_count = len(reference_ngrams)
    generated_ngram_count = len(generated_ngrams)
    # Check for zero division
    if generated_ngram_count == 0 or reference_ngram_count == 0:
        cider_score = 0.0  # Assign a score of 0 when dividing by zero
    else:
        cider_score = intersection_ngram_count / (generated_ngram_count * reference_ngram_count) ** 0.5
    #print(cider_score)
    if cider_score >= 0.15:
        cider_value.append(cider_score)
        cnt += 1
sum_cider =sum(cider_value)

print("Total caption is : ",cnt)
print(f"Avarage CIDEr score: {sum_cider/cnt}")
```
